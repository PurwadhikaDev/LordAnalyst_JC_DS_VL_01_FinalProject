# **4. Conclusion and Recommendation**

## 4.1 Conclusion

> Kita akan mebagi beberapa poin kesimpulan dari apa yang sudah kita kerjakan sejauh ini menjadi beberapa bagian sebagai berikut :
1. Pada bagian EDA, 
ditemukan bahwa  Beberapa kolom dimana jumlah salah satu value mendominasi ( >60% ) adalah **'Coverage' , 'Employment Status', 'Location Code' , 'Policy Type' dan ' Vehicle Size'**. Terdapat hubungan yang kuat antara total claim amount dengan kolom (location code dan Employment Status) sementara Monthly Premium Auto terdapat hubungan yang kuat dengan Vehicle Class. Apabila akan memprediksi kedua fitur tersebut maka kolom tersebut direkomendasikan sebagai variable independent pada model machine learning. 
2. Kita sudah **membuat model regresi untuk memprediksi jumlah premi perbulan calon nasabah baru** dengan menggunakan dataset yang kita punya. Model regresi yang kita pilih adalah menggunakan model regresi non-linear dengan Random Forest yang sudah kita tuning paraemeternya sehingga menghasilkan **score MAE pada data test yaitu sebesar 11.90** (skor MAE terkecil). Adapun fitur yang memiliki kontribusi yang cukup besar terhdadap penentuan jumlah premi calon nasabah baru adalah **fitur 'Vehicle Class'** dengan total nilai **feature importance sebesar 83%**.  
3. Setelah kita membuat model prediksi jumlah premi, kita juga telah **membuat model regresi yang dapat memprediksi jumlah total klaim asuransi calon nasabah baru**. Model prediksi total klaim asuransi ini menggunakan model regresi non-linear dengan XGBoost yang sudah kita tuning parameternya sehingga menghasilkan **score MAE pada data test yaitu sebesar 72.12** (skor MAE terkecil). Adapun fitur yang memiliki kontribusi yang cukup besar dalam penentuan total klaim asuransi nasabah adalah nasabah yang tinggal di pinggiran kota (suburban) dan pedesaan (rural), lalu ada jumlah premi perbulan nasabah, dan status perkawinan yang masih lajang (single). **Keempat variabel tersebut menyumbang feature importance sebesar 97.1% dari seluruh variabel**.

## 4.2 Business Impact  
> Berdasarkan kesimpulan diatas, kita dapat menjabarkan dampak bisnis apa yang akan terjadi dari sisi perusahaan.  
1. **Biaya tenaga kerja** untuk karyawan yang menganalisa penentuan jumlah premi perbulan calon nasabah baru dapat kita alihkan untuk biaya operasional lain yang lebih dibutuhkan oleh perusahaan.
1. **Proses registrasi calon nasabah baru akan lebih cepat** karena perusahaan kini sudah memiliki model yang dapat menentukan jumla premi perbulan calon nasabah baru dengan cepat.
1. **Perusahaan dapat menerapkan beberapa langkah preventif** untuk menangani calon nasabah baru yang diprediksi akan memiliki total klaim asuransi yang tinggi di masa depan.
1. **Perusahaan juga memiliki opsi** apakah ingin memberi semacam diskon kepada calon nasabah yang memiliki proyeksi total klaim asuransinya rendah di masa depan agar **calon nasabah tersebut tertarik untuk mengasuransikan kendaraannya kepada perusahaan**.

## 4.3 Recommendation

> Rekomendasi yang dapat kami sampaikan adalah sebagai berikut:  
1. Untuk calon nasabah yang memiliki prediksi total klaim asuransi yang tinggi di masa depan, **perusahaan dapat memberikan semacam pelatihan khusus tentang keselamatan berkendara, dan juga tips and trick merawat kendaraan dengan baik**. Seperti kata pepatah bilang "Mencegah lebih baik daripada mengobati". Kita dapat mencegah hal-hal yang tidak kita inginkan di masa depan dan menjauhi segala macam hal-hal yang dapat mengundang kecelakaan dan musibah kepada para nasabah di masa depan.
1. Keterbatasan data yang kita dapatkan untuk mengerjakan model ini seharusnya dapat menjadi perhatian bagi pihak perusahaan. **Pihak perusahaan dapat melakukan investasi lebih terhadap ketersediaan data perusahaan** agar proses analisa data dapat lebih komprehensif. Apabila perusahaan sudah berada pada tahap "Data Driven", perusahaan sudah bisa mengumpulkan, menganalisis, meramalkan, hingga mengautomasi proses 
yang mereka miliki menggunakan data.
1. Model prediksi yang sudah kita buat masih dapat disempurnakan kembali oleh perusahaan. Kita menyadari **keterbatasan data, waktu, dan tenaga yang kita curahkan untuk membuat model prediksi ini dengan segala sumber daya yang kita miliki**. Kedepannya, kami akan sangat menghargai apabila model ini dapat disempurnakan menjadi lebih baik lagi dan memberikan dampak yang lebih baik lagi kepada perusahaan maupun nasabah asuransi kendaraan.
