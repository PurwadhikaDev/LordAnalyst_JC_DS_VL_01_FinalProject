# 2. Problem Business

> **Perusahaan "Lord Insurance Auto" bergerak di bidang jasa asuransi mobil yang beroperasi di Negara Amerika Serikat.** Perusahaan ini memiliki sejumlah nasabah yang telah mengasuransikan mobilnya kepada perusahaan. Premi adalah sejumlah uang yang ditetapkan oleh perusahaan asuransi dan disetujui oleh pemegang polis untuk dibayarkan berdasarkan perjanjian asuransi. Premi dalam jasa asuransi mobil ini, dibayarkan setiap sebulan sekali kepada perusahaan asuransi oleh para nasabah sebagai ganti mendapat jaminan terhadap mobil yang diasuransikannya.  

> Perusahaan Lord Insurance Auto berhak menentukan jumlah premi yang harus dibayarkan oleh calon nasabah setiap bulannya. Penentuan premi ini sangat berpengaruh terhadap cash flow perusahaan sehingga perlu adanya pengoptimalan terhadap penentuan jumlah premi perbulan yang harus dibayar oleh calon nasabah. Keadaan perusahaan saat ini, tidak ditemukannya model yang dapat memprediksi jumlah premi yang harus dibayar oleh calon nasabah menyebabkan proses administrasi nasabah baru mengalami hambatan. Butuh waktu yang sangat lama untuk menghitung jumlah premi calon nasabah baru dan perhitungan tersebut dilakukan manual satu per satu oleh karyawan perusahaan. Proses ini dinilai tidak efektif dalam hal tingkat efisiensi kerja karyawan karena jenis pekerjaan yang sama terus dilakukan berulang-ulang sehingga membuang waktu kerja karyawan.  

> Selain mengenai perhitungan jumlah premi perbulan calon nasabah baru, perusahaan ini juga tidak memilki klasifikasi terhadap nasabah yang memiliki jumlah total klaim asuransinya yang sangat tinggi dan cenderung merugikan perusahaan. Pengklasifikasian nasabah perusahaan sangatlah penting dalam dunia bisnis, karena dengan karakteristik nasabah yang berbeda-beda kita dapat memprediksi kejadian di masa depan dan melakukan tindakan preventif untuk mencegah hal yang tidak diinginkan. Klasifikasi nasabah juga dapat digunakan kepada calon nasabah baru untuk memprediksi jumlah total klaim asuransi mereka di masa yang akan datang. Pengklasifikasiaan dapat dibagi menjadi tiga kelas yaitu ‘Low Risk’, ‘Medium Risk, dan ‘High Risk’ dengan fokus utama pada kelas ‘High Risk’.  

> Sejauh ini, dapat kita simpulkan **2 permasalahan bisnis** yang ada di perusahaan Lord Insurance Auto, yaitu sebagai berikut:
1. **Tidak adanya model perhitungan untuk menentukan jumlah premi calon nasabah baru di perusahaan Lord Insurance Auto.**
2. **Tidak adanya klasifikasi calon nasabah baru mengenai jumlah total klaim asuransi di perusahaan Lord Insurance Auto.** 

> Adapun tujuan/target dari penyelesaian masalah diatas adalah sebagai berikut:
1. **Menemukan model perhitungan yang dapat memprediksi besaran jumlah premi calon nasabah baru.**
2. **Menemukan model yang dapat mengklasifikasikan calon nasabah baru mengenai jumlah total klaim asuransi.**

## 3.2 Features Information

Penjelasan mengenai fitur yang ada di dalam dataset :
1. **Customer** : Kode unik setiap nasabah (kategorik)
2. **State** : Tempat negara bagian nasabah tinggal (kategorik)
3. **Customer Lifetime Value** : Prediksi perhitungan pendapatan perusahaan yang dapat dikumpulkan dari setiap nasabah (numerik)
4. **Response** : Respon nasabah terhadap tawaran pembaharuan polis asuransi (kategorik)
5. **Coverage** : Jenis perlindungan asuransi yang dimiliki nasabah (kategorik)
6. **Education** : Latar belakang pendidikan nasabah (kategorik)
7. **Effective To Date** : Tanggal polis asuransi mulai berlaku (date)
8. **EmploymentStatus** : Status pekerjaan nasabah (kategorik)
9. **Gender** : Jenis kelamin nasabah (kategorik)
10. **Income** : Pendapatan perbulan nasabah (numerik)
11. **Location Code** : Jenis wilayah tempat tinggal nasabah (kategorik)
12. **Marital Status** : Status perkawinan nasabah (kategorik)
13. **Monthly Premium Auto** : Jumlah premi perbulan yang harus dibayar nasabah (numerik)
14. **Months Since Last Claim** : Jumlah bulan saat terakhir mengajukan klaim asuransi (numerik)
15. **Months Since Policy Inception** : Jumlah bulan saat terakhir polis asuransi berlaku atau diperbaharui oleh kedua pihak (numerik)
16. **Number of Open Complaints** : Jumlah komplen yang diajukan oleh nasabah kepada perusahaan asuransi (numerik)
17. **Number of Policies** : Jumlah polis yang dimiliki nasabah (numerik)
18. **Policy Type** : Tipe polis yang dimiliki nasabah (kategorik)
19. **Policy** : Jenis polis yang dimiliki nasabah (kategorik)
20. **Renew Offer Type** : Jenis tawaran pembaharuan yang diajukan kepada nasabah (kategorik)
21. **Sales Channel** : Jenis saluran pemasaran setiap nasabah (kategorik)
22. **Total Claim Amount** : Besaran jumlah total klaim asuransi yang telah diajukan oleh nasabah (numerik)
23. **Vehicle Class**	: Jenis kelas kendaraan yang dimiliki nasabah (kategorik)
24. **Vehicle Size** : Jenis besar kendaraan yang dimiliki nasabah (kategorik)
